### 3.12

Suppose that $X$ is a random variable taking values in {1,2,3,...} with probability mass function

$p_X(n) = \frac{6}{\pi^2} \cdot \frac{1}{n^2}$

Show that $E[X] = \infty$

Solution: 

$E[X] = \Sigma_{n = 1}^{\infty}n ( \frac{6}{\pi^2} \cdot \frac{1}{n^2})$

$=\frac{6}{\pi^2} \Sigma_{n=1}^{\infty}\frac{1}{n}$

$= \frac{6}{\pi^2} [1+\frac{1}{2}+\frac{1}{3}+...]$

$ = E[X] = \infty$
